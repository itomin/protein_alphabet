```python
from Bio.Seq import Seq
from Bio import SeqIO

def readPDB(path):
    for record in SeqIO.parse(path, "fasta"):
        pdb = "".join(list(record.seq))
    return pdb
```

```python
import pandas as pd

df = pd.read_csv("pb_hits_sample.csv")
df = df[df.hit_length >= 10]
```
