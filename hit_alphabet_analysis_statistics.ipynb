```python
pb = pd.read_csv("../data/pdb_str/statistics/pb_histogram_count_by_hit.csv")
m32k25 = pd.read_csv("../data/pdb_str/statistics/m32k25_histogram_count_by_hit.csv")
chem = pd.read_csv("../data/pdb_str/statistics/chem_histogram_count_by_hit.csv")
cs219 = pd.read_csv("../data/pdb_str/statistics/cs219_histogram_count_by_hit.csv")
aa = pd.read_csv("../data/pdb_str/statistics/aa_histogram_count_by_hit.csv")
```
